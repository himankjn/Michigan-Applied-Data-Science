The data you have been given is near **Ann Arbor, Michigan, United States**, and the stations the data comes from are shown on the map below.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
#convert temperature from tenths of degree C to degree C
df = pd.read_csv('weather.csv')
df['Data_Value']=0.1*df.Data_Value
days=list(map(lambda x: x.split('-')[-2]+'-'+x.split('-')[-1], df.Date))
years=list(map(lambda x: x.split('-')[0], df.Date))
df['Days']=days 
df['Years']=years
df_2005_to_2014=df[(df.Days!='02-29')&(df.Years!='2015')]
df_2015=df[(df.Days!='02-29')&(df.Years=='2015')]
df_max=df_2005_to_2014.groupby(['Element','Days']).max()
df_min = df_2005_to_2014.groupby(['Element','Days']).min()
df_2015_max=df_2015.groupby(['Element','Days']).max()
df_2015_min = df_2015.groupby(['Element','Days']).min()
maxtill2014=df_max.loc['TMAX'].Data_Value
mintill2014=df_min.loc['TMIN'].Data_Value
max2015=df_2015_max.loc['TMAX'].Data_Value
min2015=df_2015_min.loc['TMIN'].Data_Value
x=pd.date_range('1900-01-01',periods=365)
```

```python
from matplotlib import dates
plt.figure(figsize=(10,5))
plt.plot(x,mintill2014,'-',label='Minimum Temperature')
plt.plot(x,maxtill2014,'-',label='Maximum Temperature')
plt.fill_between(x,y1=mintill2014,y2=maxtill2014,alpha=0.1,color='orange')
```

```python
recordhigh=max2015[max2015>maxtill2014]
recordlow=min2015[min2015<mintill2014]
recordhigh.index=pd.to_datetime(recordhigh.index,format='%m-%d')
recordlow.index=pd.to_datetime(recordlow.index,format='%m-%d')
```

```python

plt.scatter(recordlow.index,recordlow.values,label='record low')
plt.scatter(recordhigh.index,recordhigh.values,label='record high')
plt.subplots_adjust(bottom=0.2)
plt.gca().xaxis.set_major_formatter(dates.DateFormatter('%b'))
plt.legend()
plt.title('New Orelands Record Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature in $^{\circ}$C')
plt.savefig('temp.png')   
```
