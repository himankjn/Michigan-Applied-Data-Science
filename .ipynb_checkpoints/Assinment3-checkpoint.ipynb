```python
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
mean = df.mean(axis=1).tolist()
std = df.std(axis=1)
n=df.shape[1]
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib notebook
```

```python
df=df.T
df.head()
```

```python

n=df.shape[0]
std=df.std(axis=0).values.tolist()
se=std/np.sqrt(n)
y=df.mean(axis=0)
colm=plt.cm.get_cmap('Blues')
yabs=39500
diff=y-min(y)
cols=diff/max(diff)
```

```python

plt.ylim(30000,50000)
plt.bar(range(len(df.columns)),y,tick_label=list(df.columns),color=colm(cols),yerr=se,capsize=15)
plt.axhline(y=yabs,color='grey')
sm=cm.ScalarMappable(norm=plt.Normalize(0,max(cols)),cmap=colm)
sm.set_array([])
cbar=plt.colorbar(sm,ticks=np.arange(0,1.2,0.2))
cbar.ax.set_yticklabels(np.linspace(30000,50000,6))
```
