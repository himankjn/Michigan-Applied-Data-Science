```python
import pandas as pd
import numpy as np
#https://www.kaggle.com/abecklas/fifa-world-cup#WorldCupMatches.csv
#https://www.kaggle.com/abecklas/fifa-world-cup#WorldCups.csv
wc=pd.read_csv('WorldCups.csv')
wc['Attendance']=list(map(lambda x: int(x.replace('.','')),wc['Attendance']))
wc.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(10,6),sharex=True,sharey=True)
ax1.plot(wc['Year'],wc['QualifiedTeams']/10,label="No. of Teams Qualified in 10's")
ax2.plot(wc['Year'],wc['GoalsScored']/100,label="Goals Scored in 100's")
ax3.plot(wc['Year'],wc['MatchesPlayed']/10,label="No. of Matches Played in 10's")
ax4.plot(wc['Year'],wc['Attendance']/1000000,label='Attendance in millions')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
fig.suptitle("WorldCup Trends")
```

```python
import matplotlib.cm as cm
import seaborn as sns
countrywins=wc.groupby('Winner')['GoalsScored'].agg(len).sort_values()
col=countrywins.values/countrywins.values.max()
colm=plt.cm.get_cmap('RdPu')
plt.figure(figsize=(8,4))
plt.bar(countrywins.index,countrywins.values,color=colm(col))
plt.xticks(rotation=45)
plt.subplots_adjust(bottom=0.2)
sm = cm.ScalarMappable(cmap=colm, norm=plt.Normalize(0,max(countrywins.values)))
sm.set_array([])
plt.colorbar(sm)
```

```python
df=pd.read_csv('WorldCupMatches.csv')
brazilhome=df[df['Home Team Name']=='Brazil']['Home Team Goals'].mean()
othershome=df[df['Home Team Name']!='Brazil']['Home Team Goals'].mean()
brazilaway=df[df['Away Team Name']=='Brazil']['Away Team Goals'].mean()
othersaway=df[df['Away Team Name']!='Brazil']['Away Team Goals'].mean()
brazil=[brazilhome,brazilaway]
others=[othershome,othersaway]
```

```python
plt.figure()
plt.style.use('seaborn-dark-palette')
x=np.arange(2)
plt.bar(x,[brazilhome,brazilaway],width=0.35,label='Brazil')
plt.bar(x+0.35,[othershome,othersaway],width=0.35,label='Other Countries')
plt.xticks(ticks=x+0.35/2,labels=['Home','Away'])
plt.yticks([])
plt.legend();
plt.title('Brazil vs Other Countries Home/Away Goals')
for i,v in enumerate(brazil):
    plt.text(i-0.05,v+0.01,str(round(v,2)),color='black',fontweight='bold')
for i,v in enumerate(others):
    plt.text(i+0.27,v+0.01,str(round(v,2)),color='black',fontweight='bold')
```
